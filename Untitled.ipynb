```python
import pandas as pd
import glob 
import seaborn as sns
import matplotlib.pyplot as plt
# import geopandas as gp
from flask import Flask 

```

```python
paths = glob.glob("data/*.csv")
dfs = map(pd.read_csv,paths)
df = pd.concat(dfs)

df
```

```python
plt.figure(figsize=(9,18))
sns.displot(data=df, x="realSum",kind="kde")
```

```python
plt.figure(figsize=(2,3))
sns.boxplot(data=df,y="realSum") 
```

```python
plt.figure(figsize=(20,8))
ax=sns.stripplot(data=df.head(100), x="lng",y="lat",hue="room_type",palette="Dark2")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax=sns.scatterplot(data=df.head(1000), x="lng",y="lat",hue="room_type",palette="Dark2",size="metro_dist",sizes=(1,1000))
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,8))
ax=plt.pie(df["room_type"].value_counts(),autopct="%1.1f%%",explode=[0.2, 0.2,0.1])
plt.show()
```

# Flask-it!

```python
from flask import Flask, render_template
```

```python
app = Flask("app", template_folder='template', static_folder="static") 
# app.run()
```

```python
@app.route("/<i>")
def getbyid(i):
    return df.iloc[int (i)].to_dict()
```

```python
@app.route("/info")
def getinfo():
    return render_template("info.HTML")
```

```python
@app.route("/house/<i>")
def getbyidhtml(i):
    k = df.iloc[int (i)].to_dict()
    return render_template("info.HTML", imane=k)
```

```python
getbyid(5)
```

```python
app.run()
```
